```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial_atualizado_2.csv', sep = ';')
```

```python
dados.head(10)
```

```python
dados['Valor'].mean()
```

 #### https://pandas.pydata.org/pandas-docs/stable/api.html#api-dataframe-stats

```python
bairros = ['Barra da Tijuca', 'Botafogo', 'Copacabana', 'Flamengo', 'Ipanema', 'Leblon', 'Tijuca']
```

```python
selecao = dados['Bairro'].isin(bairros)
```

```python
dados = dados[selecao]
```

```python
dados
```

```python
dados['Bairro'].drop_duplicates()
```

```python
dados
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data.Valor.mean()))
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

# Estatística Descritivas

```python
grupo_bairro['Valor'].describe()
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max'])
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min' : 'Minimo', 'max' : 'Maximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 30))
```

```python
grupo_bairro['Valor'].std().plot.bar(color = 'blue')
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```
