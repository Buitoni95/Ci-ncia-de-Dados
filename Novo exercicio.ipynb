```python
import pandas as pd
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 30))
```

```python
dados = pd.read_csv('dados/aluguel_residencial_atualizado_2.csv', sep = ';')
```

```python
bairros = ['Barra da Tijuca', 'Botafogo', 'Copacabana', 'Flamengo', 'Ipanema', 'Leblon', 'Tijuca']
```

```python
selecao = dados['Bairro'].isin(bairros)
```

```python
dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()
```

```python
dados
```

```python
grupo_bairros_para_iptu = dados.groupby('Bairro')
```

```python
grupo_bairros_para_iptu.groups
```

```python
grupo_bairros_para_iptu[['IPTU', 'Valor m2']].mean().round(2)
```

```python
grupo_bairros_para_iptu['IPTU'].describe()
```

```python
grupo_bairros_para_iptu['IPTU'].aggregate(['min', 'max']).rename(columns = {'min' : 'Minimo', 'max' : 'Maximo'})
```

```python
fig = grupo_bairros_para_iptu['IPTU'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do IPTU')
fig.set_title('Valor Médio do IPTU nas Regiões abaixo', {'fontsize': 22})
```

```python
precos = pd.DataFrame([['Feira', 'Cebola', 2.5], 
                        ['Mercado', 'Cebola', 1.99], 
                        ['Supermercado', 'Cebola', 1.69], 
                        ['Feira', 'Tomate', 4], 
                        ['Mercado', 'Tomate', 3.29], 
                        ['Supermercado', 'Tomate', 2.99], 
                        ['Feira', 'Batata', 4.2], 
                        ['Mercado', 'Batata', 3.99], 
                        ['Supermercado', 'Batata', 3.69]], 
                        columns = ['Local', 'Produto', 'Preço'])
precos
```

```python
produtos = precos.groupby('Produto')
produtos.describe().round(2)
estatisticas = ['mean', 'std', 'min', 'max']
estatisticas_nomes = {'mean': 'Média', 'std': 'Desvio Padrão', 'min': 'Mínimo', 'max' : 'Máximo'}
```

```python
produtos['Preço'].aggregate(estatisticas).rename(columns = estatisticas_nomes).round(2)
```
